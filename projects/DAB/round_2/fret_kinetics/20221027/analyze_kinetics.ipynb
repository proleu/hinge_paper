```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set(
    context="notebook",
    style="ticks",  # make the background white with black lines
    palette="colorblind",  # a color palette that is colorblind friendly!
)

from scipy.optimize import curve_fit
def order_1_curve(t, a, k, start): # units of luminescence, t = time (s), a = saturation signal, k = rate constant (1/s)
    return a * (1 - np.exp(-k * t)) + start

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6,6))

t_data = pd.to_timedelta(binding_df.loc[:, "Time"].astype(str)).dt.total_seconds()

fits = {}

binding_data = binding_df.iloc[:160, [i*12+2 for i in range(8)]]
# binding_data = np.where(binding_data == "OVRFLW", 6000000, binding_data)
for i, color in zip(range(8), [sns.color_palette("flare", as_cmap=True)(x) for x in np.linspace(0, 1, 8)]):
    s0 = start_conc*(i+1)/8
    binding_col = binding_data.iloc[:, i]
    max_i = np.argmax(binding_col)
    binding_fit = binding_col[:max_i]
    t_fit = t_data[:max_i]
    sns.scatterplot(x=t_data, y=binding_col, label=f'{s0/1000000000:.00e} M', zorder=-1, ax=ax, color=color)
    try:
        (sat, k_app, start), pcov = curve_fit(order_1_curve, t_fit, binding_fit, bounds=((0,0,0),(np.inf,0.1,np.inf)))
        sns.lineplot(x=t_fit, y=order_1_curve(t_fit, sat, k_app, start), color=color, ax=ax)
        fits[s0] = (sat, k_app, start)
        print(f'{s0/1000000000:.01e} M: {sat:.02e}, {k_app:.02e}, {start:.02e}')
    except RuntimeError:
        print(f"fit {i} failed")
ax.set_xlabel(f'Time (s)')
ax.set_ylabel('Ex 520nm / Em 665nm (a.u.)')
ax.legend()

fig.tight_layout()
fig.patch.set_alpha(0)
fig.savefig("CS221_DAB21_kinetics_fit.png", dpi=300)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6, 6))

sns.scatterplot(x=[c/1000000000 for c in fits.keys()], y=[fit[1] for fit in fits.values()], ax=ax)

ax.set_xlabel(f'[DAB21] (M)')
ax.set_ylabel('k_app (1/s)')

fig.tight_layout()
fig.patch.set_alpha(0)
fig.savefig("CS221_DAB21_kinetics_kapp_conc.png", dpi=300)
```
